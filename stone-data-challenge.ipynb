#### **Descrição do case**

A Stone tem como objetivo ajudar o empreendedor brasileiro a vender, gerir e crescer seu negócio através dos nossos produtos e serviços que vão desde a maquininha de cartão de crédito à sistemas ERP para gestão.

Um dos produtos que oferecemos aos nossos clientes é o crédito, voltado para facilitar a vida do lojista e impulsionar o crescimento do seu negócio.

A Stone trabalha desde 2019 para construir um produto sólido e maduro.

Mesmo no momento mais intenso da pandemia provocada pela COVID-19, em 3 anos, já liberamos mais de 190 mil linhas de crédito, que somam mais de 3,4 bilhões de reais desembolsados para apoiar nossos clientes a crescer e manter seus negócios.

Nosso produto é diferenciado: os clientes pagam seus empréstimos através de um percentual de retenção aplicado sobre as transações realizadas pela maquininha de cartão Stone.

Desta forma, conseguimos promover um pagamento sustentável do empréstimo que acompanha as variações de fluxo de caixa do lojista, que não precisa se comprometer com um valor fixo que possa pesar em um mês de baixo movimento em seu negócio.

Um cliente pode ter mais de um contrato de empréstimo simultâneo, desde que o contrato anterior esteja em dia.

Durante a pandemia, naturalmente tivemos um aumento de clientes inadimplentes, ou seja, com problemas no pagamento de seus contratos e, consequentemente, sentimos a necessidade de evoluir nossas ações de comunicação e acionamento para recuperação dos saldos devedores que alguns clientes passaram a deixar em aberto.

Dentro desta evolução, uma das ações tomadas internamente foi refinar nossas estratégias de comunicação com os clientes, visando recuperar clientes que apresentavam problemas no ritmo de pagamento, e claro, sempre usando dados para melhor guiar nossas decisões!

#### **Como funciona os Acionamentos?**


Quando um cliente começa a apresentar dificuldade na liquidação de seus contratos, ele passa a ser elegível para ser acionado por nossa régua de comunicação com intuito de estimular a retomada do ritmo saudável de pagamento de acordo com o perfil e o momento de cada cliente.

As comunicações são disparadas por contrato, sendo assim, um cliente pode receber diversas comunicações com diferentes conteúdos, um para cada momento da jornada de seu respectivo contrato. Desde o envio da comunicação até o recebimento, são gerados flags que indicam o engajamento do cliente com as comunicações da seguinte forma:

> 1.Envio do comunicado > 2.Comunicado recebido > 3.Comunicado lido

Dada a sequência da comunicação acima, ocasionalmente são observados erros no processo de envio do comunicado, gerando um percentual (%) de falhas por diversos motivos (servidor indisponível, número incorreto etc.). Após o envio bem-sucedido, observamos também que nem todos os comunicados são lidos.

Além do engajamento do cliente com as comunicações, temos outra questão importante: a efetividade na recuperação dos contratos ativos. Uma comunicação eficaz é aquela que estimula e converte um cliente com problemas na liquidação do seu contrato a retomar o pagamento saudável, seja por meio de quitação ou do retorno ao ritmo de transação esperado.

Para mensurar a efetividade de uma ação, temos dois ótimos termômetros: os dias sem pagamento (dsp) e os dias sem pagamento do principal de um cliente (dspp).

Os dias sem pagamento (dsp) representam o total de dias corridos que um contrato apresenta sem realizar nenhum pagamento.

Já os dias sem pagamento do principal (dspp), representam o total de dias corridos que um contrato apresenta sem reduzir o valor do saldo principal. Neste último conceito, mesmo que o contrato apresente algum pagamento, se este montante não for suficiente para cobrir juros + impostos, valores deduzidos prioritariamente, o saldo principal do contrato permanecerá sem pagamento.

Consideramos um diferencial a exploração de variáveis complementares (ex. ramo do lojista, dados geográficos, sazonalidade etc.) para entender se existem diferentes curvas para diferentes perfis, visando customizar as ações de comunicação de acordo com o perfil de cada cliente.

Estas análises extras também podem (e devem) se estender a performance de pagamento e efetividade das ações de cobrança.

# 2. Funções

```python
# Função para verificar o tamanho dos datasets
def check_size_dataset(data):
    dataset = os.path.getsize('datasets/' + data + '.csv')/1024/1024
    return dataset

# Função para carregar os datasets
def read_csv(data, n_rows):
    dataset = pd.read_csv('datasets/' + data + '.csv', nrows=n_rows)
    return dataset


# Função para mostrar os tipos dos dados e as métricas de todos os datasets de forma dinâmica
def infos_data(data, subset):
    type = data.dtypes
    count = data.count()
    missing = data.isnull().sum() / len(data) * 100
    unique = data.nunique()
    duplicated = data[data.duplicated(keep=False, subset=[subset])].sort_values(by=subset).count()

    tabel = pd.concat([type, count, missing, unique, duplicated], axis = 1, keys = ['type', 'count', 'missing', 'unique', 'duplicated'])
    return tabel




# FUNÇÃO PARA CRIAR TABELA DE DISTRIBUIÇÃO DE FREQUÊNCIAS PARA DADOS QUALITATIVOS E QUANTITATIVOS DISCRETOS

# Frequência absoluta - Fi
# Frequência relativa - Fri
# Frequência acumulada - Fac
# Frequência relativa acumulada - Frac

def dist_freq(data, column):
    Fi = data[column].value_counts()
    Fri = data[column].value_counts(normalize=True)
    Fac = Fi.cumsum()
    Frac = (Fi.cumsum() / len(data[column]) * 100) 

    dist_freq_table = pd.DataFrame(
        {'Fi': Fi,
        'Fri': Fri,
        'Fac':Fac,
        'Frac':Frac}
    )  

    dist_freq_table.loc['Soma'] = (dist_freq_table[['Fi', 'Fri']].sum())
    dist_freq_table.rename_axis(column, axis='columns', inplace = True)
    # dist_freq_table.sort_index(ascending=False)
    return dist_freq_table

#  FUNÇÃO PARA CRIAR TABELA DE DISTRIBUIÇÃO DE FREQUÊNCIAS PARA DADOS CONTÍNUOS AGRUPADOS EM CLASSES
#  DETERMINAR O NÚMERO DE CLASSES (k), UTILIZANDO A EXPRESSÃO DE STURGES

def distr_freq_continuas(data, column):

    # Determinaremos o número de classes (k) pela expressão de Sturges
    k = round(1 + 3.3 * np.log10(len(data)))

    # O intervalo entre as classes (h) é dado por:
    h = round((data[column].max() - data[column].min()) / k)

    # Por fim, construiremos a tabela de distribuição de frequências para cada classe.

    # Frequência absoluta
    Fi = pd.value_counts(
        pd.cut(
            x = data[column],
            bins = h,
            include_lowest=True
        ),
        sort = False
    )

    # Frequência relativa
    Fri = pd.value_counts(
        pd.cut(
            x = data[column],
            bins = h,
            include_lowest=True
        ),
        sort = False,
        normalize=True
    )
    
    # Frequência acumulada
    Fac = Fi.cumsum()

    # Frequência relativa acumulada
    Frac = (Fi.cumsum() / len(data[column]) * 100)

    dist_freq_table = pd.DataFrame(
        {   'Fi':Fi,
            'Fri': Fri,
            'Fac':Fac,
            'Frac':Frac
        }
   )
    
    dist_freq_table.rename_axis(column, axis='columns', inplace = True)
    return dist_freq_table


# Função para mudar feature para datetime
def transformar_datetime(dataset, lst_colunas, formato):
    for coluna in lst_colunas:
        dataset[coluna] = pd.to_datetime(dataset[coluna], format = formato, errors='coerce')
        

# Função para calcular a média aritmética simples
def media_aritmetica(data, column):
    x = data[column].sum() / len(data[column])
    return x

# Função para calcular a média aritmética ponderada
def media_aritmetica_pond(data, column_1, column_2):
    x = (data[column_1] * data[column_2]).sum() / data[column_2].sum()
    return x 

# Função para calcular a média aritmética para dados discretos agrupados
def media_aritmetica_disc_agrup(data, column):
    x = (data[column] * data[column].value_counts()).sum() / len(data)
    return x

# Função para criar tabela de estatística descritiva
def table_describe(data_att):
    # Medidas de Tendência Central
    ct1 = pd.DataFrame(data_att.apply(np.mean)).T
    ct2 = pd.DataFrame(data_att.apply(np.median)).T


    # Medidas de Dipersão
    d1 = pd.DataFrame(data_att.apply(np.std)).T
    d2 = pd.DataFrame(data_att.apply(min)).T
    d3 = pd.DataFrame(data_att.apply(max)).T
    d4 = pd.DataFrame(data_att.apply(lambda x: x.max() - x.min())).T
    d5 = pd.DataFrame(data_att.apply(lambda x: x.skew())).T
    d6 = pd.DataFrame(data_att.apply(lambda x: x.kurtosis())).T

    # concat
    m = pd.concat([d2, d3, d4, ct1, ct2, d1, d5, d6]).T.reset_index()
    m.columns = ['attributes', 'min', 'max', 'amplitude', 'mean', 'median', 'std', 'skew', 'kurtosis']
    round(m,2)
    return m

# Função para detectar outliers univariados moderados
def detect_outliers_univar_moderados(data, column):

    q1 = np.quantile(data[column], 0.25)
    q3 = np.quantile(data[column], 0.75)
    lower_range = q1 - 1.5 * (q3 - q1)
    upper_range = q3 + 1.5 * (q3 - q1)

    if lower_range < q1 or upper_range > q3:
        print('Q1: {}, Q3: {}'.format(q1, q3))
        print('Lower Range: {}, Upper Range: {}, Há Outlier Moderado'.format(lower_range, upper_range))
    else:
        print('Q1: {}, Q3: {}'.format(q1, q3))
        print('Lower Range: {}, Upper Range: {}, Não Há Outlier Moderado'.format(lower_range, upper_range))

# Função pra gerar o coeficiente de variação
def coef_variacao(data, column):
    cv = (data[column].std() / data[column].mean()) * 100
    if cv > 30:
        print('Porcentagem: {}, A média do conjunto de dados não é uma boa medida para representação dos dados.'.format(cv))
    else:
         print('Porcentagem: {}, A média do conjunto de dados é uma boa medida para representação dos dados.'.format(cv))


```

- **flag_transacao**: Flag indicando se, na data de referência, houve algum valor transacionado na maquininha do cliente (TPV).
